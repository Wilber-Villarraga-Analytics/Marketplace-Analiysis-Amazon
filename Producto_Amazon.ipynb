Vamos a convertir cada columna al tipo de dato adecuado para facilitar el análisis de la información. En el caso de la columna best_sellers_rank, la separaremos en dos columnas: una que contenga únicamente la posición (como valor numérico) y otra que contenga la categoría. #### best_sellers_rank separar en columnas de numero y categoria

```python
# cambiar "#" 
df_filtrado["best_sellers_rank"] = (
    df_filtrado["best_sellers_rank"] .str.replace("#", "", regex=False)
)

# Separar
df_filtrado[["rank_position", "rank_category"]] = (
    df_filtrado["best_sellers_rank"].str.split(" ", n=1, expand=True)
)

# Convertir posición a número entero
df_filtrado["rank_position"] =pd.to_numeric(
    df_filtrado["rank_position"],errors="coerce").astype("Int64")

# Remplazar valores incoerentes
df_filtrado.loc[df_filtrado["rank_category"] == "encontrado", "rank_category"] = "no encontrado"

# Eliminar columna 
df_filtrado = df_filtrado.drop(columns=["best_sellers_rank"])
```

#### recent_purchases: separar y dejar el numero con unidades numericas.  

```python
# 1 Extraer solo el número antes del "+"
df_filtrado["recent_purchases"] = (
    df_filtrado["recent_purchases"].str.extract(r"(\d+)\+", expand=False)
)

# 2 Remplazar "K" por "000" para representar unidades de mil 
df_filtrado['recent_purchases'] = (
    df_filtrado["recent_purchases"] .str.replace("K", "000")
)

# 3 Convertir a número entero
df_filtrado["recent_purchases"] = pd.to_numeric(
    df_filtrado["recent_purchases"],
    errors="coerce"
)
# 4 Comvertir los valores nulos en 0
df_filtrado["recent_purchases"] = df_filtrado["recent_purchases"].fillna(0)
```

#### rating_stars: convertir en numero decimal (float64)

```python
# 1. Extraer todo antes del primer espacio
df_filtrado["rating_stars"] = df_filtrado["rating_stars"].str.extract(r"^(\S+)", expand=False)

# 2. Convertir a número flotante
df_filtrado["rating_stars"] = df_filtrado["rating_stars"].astype(float)
```

#### rating_count: convertir en numero entero (int64)

```python
# 1 Extraer todo antes del primer espacio
df_filtrado["rating_count"] = df_filtrado["rating_count"].str.extract(r"^(\S+)", expand=False)

# 2 Convertir "," en "."
df_filtrado["rating_count"] = (
    df_filtrado["rating_count"] .str.replace(",", "", regex=False)
)

# 2 Convertir a número entero
df_filtrado["rating_count"] = pd.to_numeric(
    df_filtrado["rating_count"],
    errors="coerce")
```

#### rating_distribution/(1,2,3,4,5)star: convertir en numero entero (int64)


```python
# rating_distribution/1star
# 1 Extraer solo el número antes del "%" 
df_filtrado["rating_distribution/1star"] = (
    df_filtrado["rating_distribution/1star"].str.extract(r"(\d+)\%", expand=False)
)
# 2 convertir en numero entero
df_filtrado["rating_distribution/1star"] =  pd.to_numeric(df_filtrado["rating_distribution/1star"])

# rating_distribution/2star
# 1 Extraer solo el número antes del "%" 
df_filtrado["rating_distribution/2star"] = (
    df_filtrado["rating_distribution/2star"].str.extract(r"(\d+)\%", expand=False)
)
# 2 convertir en numero entero
df_filtrado["rating_distribution/2star"] =  pd.to_numeric(df_filtrado["rating_distribution/2star"])

# rating_distribution/3star
# 1 Extraer solo el número antes del "%" 
df_filtrado["rating_distribution/3star"] = (
    df_filtrado["rating_distribution/3star"].str.extract(r"(\d+)\%", expand=False)
)
# 2 convertir en numero entero
df_filtrado["rating_distribution/3star"] =  pd.to_numeric(df_filtrado["rating_distribution/3star"])

# rating_distribution/4star
# 1 Extraer solo el número antes del "%" 
df_filtrado["rating_distribution/4star"] = (
    df_filtrado["rating_distribution/4star"].str.extract(r"(\d+)\%", expand=False)
)
# 2 convertir en numero entero
df_filtrado["rating_distribution/4star"] =  pd.to_numeric(df_filtrado["rating_distribution/4star"])

# rating_distribution/5star
# 1 Extraer solo el número antes del "%" 
df_filtrado["rating_distribution/5star"] = (
    df_filtrado["rating_distribution/5star"].str.extract(r"(\d+)\%", expand=False)
)
# 2 convertir en numero entero
df_filtrado["rating_distribution/5star"] =  pd.to_numeric(df_filtrado["rating_distribution/5star"])
```

```python
# Extraer el numero despues de "$"
p  = (
    df_filtrado["list_price"]
    .astype(str)
    .str.extract(r"(\d+\.\d+)", expand=False)
)

p
```
